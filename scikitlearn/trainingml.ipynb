```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python

data = pd.read_excel("iris_train.xlsx", na_values=["-"])
data = data.dropna()
```

```python
data.head() #first 5 rows of the data
```

```python
x = data.iloc[:,:-1] 
```

```python
y = data.iloc[:,-1] 
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
prediction = model.predict([[5.1,3.5,1.4,0.2]])
```

```python
prediction
```

```python
y_test=data.iloc[:,-1]
```

```python
actual = y_test
```

```python
actual
```

```python
count = 0
for i in range(0,len(actual)):
    if actual[i] == prediction:
        count += 1

print((count * 100) / len(actual))
```
