```python
import matplotlib.pyplot as plt
```

```python
years = [2016,2017,2018,2019,2020,2021,2022,2023,2024,2025]

messi_goals = [59,54,51,50,44,38,35,32,28,30]

ronaldo_goals = [55,53,44,37,36,24,18,50,35,33]

neymar_goals = [24,20,28,23,19,17,13,18,20,22]
```

plt.plot(years, messi_goals, color='blue', linestyle='--', linewidth=1, label="Messi")
plt.plot(years, ronaldo_goals, color='red', linewidth=3, label="Ronaldo")
plt.plot(years, neymar_goals, color='yellow', linewidth=2, label="Neymar")

plt.legend()
plt.grid(True)
plt.show()


```python
plt.plot(years, messi_goals, color='blue', linestyle='--', linewidth=1, label="Messi")
plt.plot(years, ronaldo_goals, color='red', linewidth=3, label="Ronaldo")
plt.plot(years, neymar_goals, color='yellow', linewidth=2, label="Neymar")
plt.title("Last 10 Years Goals of Messi, Ronaldo and Neymar")
plt.xlabel("Years")
plt.ylabel("Goals")
plt.legend()
plt.grid(True)
plt.show()
```
