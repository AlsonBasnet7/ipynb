```python
#Create a pie chart showing the market share of mobile OS (Android, iOS, others). Then recreate the same using a horizontal bar chart and observe which is easier to understand.
import matplotlib.pyplot as plt
```

```python
# Mobile OS market share data
labels = ['Android', 'iOS', 'Others']
sizes = [72, 27, 1]
colors = ['#ff9999','#66b3ff','#99ff99']
```

```python
plt.pie(sizes,
        labels=labels,
         colors=colors,
        wedgeprops={'edgecolor':'black','linewidth':1}
       )
plt.show()
```

```python
#now for the horizontal bargraph, we will be using the barh
```

```python
plt.barh(labels,sizes,color=colors)
plt.show()
```

```python
plt.barh(labels, sizes, color=colors)
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.show()
```

```python
plt.bar(labels, sizes, color=colors)
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.show()
```

```python
plt.bar(sizes,labels, color=colors)
plt.xlabel("Market Share (%)")
plt.title("Mobile OS Market Share (Bar Chart)")
plt.show()
# we interchanged the coordinated we are having the different visual of the bar, so we need to know the x and y  properly ino
```
