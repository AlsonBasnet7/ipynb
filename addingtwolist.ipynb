```python
import numpy as np
import time
```

```python
#python list
size=100000
list1=list(range(size))
list2=list(range(size))
```

```python
start = time.time()
result=[x+y for x,y in zip(list1, list2)]
end=time.time()
print("Python list addittion time:", end-start)
```

```python
#using the numpy arrya
```

```python
arr1=np.array(list1)
arr2=np.array(list2)
```

```python
start= time.time()
result=arr1+arr2 # this is called the vectorized operation
end = time.time()
print("NumPY array addition time:",end-start)
```
