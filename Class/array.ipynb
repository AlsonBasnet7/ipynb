```python
sensors=np.array([
    [12.1,14.3,23.1,11.1],
    [22.2,69.1,23.1,22.2],
    [14.2,12.2,12.2,12.2],
    [22.2,33.2,22.3,77.7]
])
    
```

```python
specific_reading=sensors[2,1]
print(specific_reading)
```

```python
top_sensors=sensors[0:2,:]
print(top_sensors)
```

```python
final_readings=sensors[:,2]
print(final_readings)
```

```python
flat_temps = np.array([22,24,35,23,23,45,46,44,32,89,12,89])
structured_temps = flat_temps.reshape(4,3)
print("The reshaped temperatures data is\n", structured_temps)
```

```python
fahrenheit_temps=(structured_temps*9/5)+32
print(fahrenheit_temps)
```

```python
#axis
students= np.array([
    [23,23,45],
    [56,78,56],
    [77,88,99]
])
print("Overall highest score in the class",np.max(students))
print("Overall average score",np.mean(students))
```

```python
subject_average = np.mean(students,axis=0)
print(subject_average)
```

```python
student_averages=np.mean(students,axis=1)
print(student_averages)
```
