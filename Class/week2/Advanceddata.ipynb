```python
import pandas as pd
import numpy as np
```

```python
df_sales= pd.read_csv("advanced_pandas_sales_dataset.csv")
```

```python
#find the total revenue generated by each product catetory
```

```python
df_sales.head()
```

```python
category_revenue= df_sales.groupby("Category")["Revenue"].sum().reset_index()
```

```python
print("__Total revenue by category__")
print(category_revenue)
```

```python
#advanced group by. Calculate multiple metrics simultaneously for each region.
```

```python
#finding the total items sold and average unit price (mean)
```

```python
region_metrics = df_sales.groupby("Region").agg(
Total_Units_Sold=("Quantity", "sum"),
Average_Item_Price=("UnitPrice", "mean")
).reset_index()
```

```python
print("\n--Advanced Metrics by Region--")
print(region_metrics)
```

```python
#here agg helps me perform sum and mean in different tables and allows me to return the output as the single dataframe
```

```python
#instead of creating separate tables for the sum and mean and then merging them, agg() does it in one step and retuns a single dataframe
```
