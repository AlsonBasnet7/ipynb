```python
#cleaning and structuring the data.
import json
```

```python
def clean_data(data):
    #removes the user
    data["users"]=[user for user in data["users"] if user["name"].strip()]
    #this removes the duplicate friends
    for user in data["users"]:
        user["friends"] = list(set(user["friends"]))
    #this removes the inactive user from the data
    data["users"]=[user for user in data["users"] if user ["friends"] or user["liked_pages"]]
    #this removes the duplicate pages
    unique_pages={}
    for page in data["pages"]:
                   unique_pages[page["id"]]=page
    data["pages"]=list(unique_pages.values())
    return data
##load, clean and display the cleaned data
data= json.load(open("codebook_data.json"))
data=clean_data(data)
josn.dump(data,open("cleaned_codeboook.json","w"),indent=4)
print("Data cleaned successfully")
    
    
```
